```python
titanic["Age"]=titanic["Age"].fillna(titanic["Age"].mean())
titanic["Survived"]=titanic["Survived"].dropna()
titanic['Embarked'].fillna(value='S', inplace=True) #most common
#titanic['Fare'].dropna(titanic['Fare'], inplace=True)
titanic.head()

#Hemos limpiado los datos que hemos considerado indispensables para el proyecto
```

ANÁLISIS MEDIANTE GRÁFICOS

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
y = titanic["Survived"].value_counts()
pie(y,autopct='%1.1f%%')
xlabel('mi grafico')
title('SUPERVIVIENTES')
legend( ('SI-NO',), loc = 'upper left')
draw()
#savefig("mi-grafico1",dpi=300)
#close()
grid(True)
show()

#Vemos que después de elimnar aquellos pasejeros que no sabemos si han sobrevivido o no, respecto a los datos de los que disponemos, un 61,4 % han sobrevivido.
```

```python
titanic['Title'] = titanic.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
titanic['Title'].unique()
```

```python
titanic['Title'] = titanic['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
titanic['Title'] = titanic['Title'].replace(['Don', 'Dona', 'Rev', 'Dr', 'Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Diferent')

plt.subplots(figsize=(13, 5))
sns.barplot(data=titanic, x='Title', y="Survived", palette="Blues_d")
plt.show()

#Las mujeres y mayores son los que más han sobrevivido.
```

```python
titanic["Age"]=titanic["Age"].astype('int')
#itanic["Age"]=titanic["Age"].round()
Age=titanic.groupby(titanic["Age"])[["Survived"]].count()
Age.plot()
#plt.scatter(Age, titanic["Survived"])
```
